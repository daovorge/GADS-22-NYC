```python
def clean_and_stem(text):
    '''This is a simple function to clean up text. This function will convert text to lower case, tokenize the test
    into words, eliminate stop words, and then stem the text'''
    
    #use the sklearn list of stop words
    stop_words = ENGLISH_STOP_WORDS
    
    #try the Lancaster stemmer - but there are others in NLTK
    stemmer = nltk.stem.LancasterStemmer()
    
    #tokenize
    tokens = nltk.word_tokenize(text.lower())
    
    #remove stop words
    clean_tokens = [word for word in tokens if word not in stop_words]
    
    #stem
    stem_tokens = [stemmer.stem(token) for token in clean_tokens]
    
    #rejoin and return
    return " ".join(stem_tokens)
```

```python
clean_training_set = [clean_and_stem(text) for text in training_data.data]
```

```python
clean_testing_set = [clean_and_stem(text) for text in testing_data.data]
```

```python
tfidf_vect = TfidfVectorizer(stop_words='english', ngram_range=(1,2))
X1_train = tfidf_vect.fit_transform(clean_training_set)
X1_test = tfidf_vect.transform(clean_testing_set)
```

```python
print X1_train.shape
print X1_test.shape
```

```python
clfLR = LogisticRegression(C=1000)
clfLR.fit(X1_train, y_train)
y1_hat = clfLR.predict(X1_test)
```

```python
print clfLR.score(X1_test, y_test)
```

```python
cm = pd.crosstab(y_test, y1_hat, rownames=["Actual"], colnames=["Predicted"])
cm
```
