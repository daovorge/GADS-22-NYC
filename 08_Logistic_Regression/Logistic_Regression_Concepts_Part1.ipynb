#CLASSIFICATION USING LINEAR REGRESSION

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

---
##Start with a two-class classification problem
---
###The output variable, $y$, takes on $1$ of $2$ values
###$y \in {0,1}$
###example: classify houses as "expensive" and "affordable". Map affordable to $0$ and expensive to $1$
##Multi-class classification problem
###example: 4 categories mapped to {0, 1, 2, 3}

---
##Why Linear Regression is not a Good Idea
---

```python
#create some simple data points
x = np.array([150, 175, 142, 287, 343, 453, 500, 398])
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])

#useful arrays to illustrate points on the plot
xx = np.linspace(100, 310, 210)
yy = np.ones(210) * 0.5

xxx = np.ones(50) * 310
yyy = np.linspace(0.0, 0.5, 50)

x1 = x
x1 = x1.reshape(len(x1),1)

#fit a linear regression model through the points
clf = linear_model.LinearRegression()
clf.fit (x1, y)

fig=plt.figure(figsize=(5,5))
ax=plt.subplot(111)
ax.set_ylim(-0.1,1.1)
ax.set_xlim(100, 550)
ax.plot(x, y, 'x')
ax.plot(x, clf.predict(x1))
ax.plot(xx, yy, color='red')
ax.plot(xxx, yyy, color='red')

ax.set_ylabel("Class: 0 = afordable, 1 = expensive")
ax.set_xlabel("House Price ($ x 1000)")
ax.set_title("Classifying Houses Using Linear Regression")
```

---
##Can we simply perform a linear regression and classify an input to class $0$ if $y \lt 0.5$, and class $1$ if $y \gt 0.5$?

```python
print clf.predict([300])
print clf.predict([310])
```
